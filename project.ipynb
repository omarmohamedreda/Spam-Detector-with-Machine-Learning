```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
# Reading the email dataset from a CSV file
raw_mail_data = pd.read_csv('mail_data.csv')
print(raw_mail_data)
```

```python
# Replacing null values in the dataset with empty strings
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)), '')
```

```python
print(mail_data.head())
```

```python
# Checking the number of rows and columns in the dataset
print(mail_data.shape)
```

```python
# Converting 'spam' to 0 and 'ham' to 1 for machine learning compatibility
mail_data.loc[mail_data['Category'] == 'spam', 'Category'] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category'] = 1
```

```python
# Separating the 'Message' column as features (X) and 'Category' column as labels (Y)
X = mail_data['Message']
Y = mail_data['Category']

print(X)
print(Y)
```

```python
# Dividing the dataset into 80% training and 20% testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)

print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
# Converting text data to numerical vectors using TF-IDF representation
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
```

```python
# Converting labels to integer type
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')

print(X_train)
print(X_train_features)
```

```python
# Initializing and training a Naive Bayes classifier with the training data
model = MultinomialNB()
```

```python
model.fit(X_train_features, Y_train)
```

```python
# Making predictions and calculating accuracy for the test set
prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
print('Accuracy on training data : ', accuracy_on_training_data)
```

```python
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
print('Accuracy on test data : ', accuracy_on_test_data)
```

```python
# Predicting whether a new email message is spam or ham
input_mail = ["I've been searching for the right words to thank you for this breather. I promise I won't take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times"]

input_data_features = feature_extraction.transform(input_mail)

prediction = model.predict(input_data_features)
print(prediction)


# Displaying the result
if prediction[0] == 1:
    print('Ham mail')
else:
    print('Spam mail')
```
